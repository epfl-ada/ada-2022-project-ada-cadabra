# Effect of sensory values on beer ratings

## Beer palate

On BeerAdvocate data:

```python
df_ratings_BA = matched_ratings.iloc[:,:17]
df_ratings_BA.set_axis(df_ratings_BA.loc[0], axis='columns', inplace=True)
df_ratings_BA = df_ratings_BA.drop(0)
df_ratings_BA.head(2)
```

```python
def getLinearRegression(attr):        
    df_ratings_BA[attr]=np.float64(df_ratings_BA[attr])
    reviews = pd.DataFrame(df_ratings_BA[[attr,'rating']]).dropna()
    x = reviews[attr]
    xmin = x.min()
    xmax = x.max()
    y = reviews['rating']
    x = x.tolist()
    y = y.tolist()
    regress = linregress(x,y)
    lr = regress
    a1 = lr[0]*xmin + lr[1]
    a2 = lr[0]*xmax + lr[1]
    return lr, a1, a2, xmin, xmax
```

```python
fig, axs = plt.subplots(2, 2)
lr, a1, a2, xmin, xmax = getLinearRegression('palate')
axs[0, 0].scatter(df_ratings_BA['palate'], df_ratings_BA['rating'], s=.5, alpha=.7, color='purple')
axs[0, 0].plot([xmin,xmax],[a1,a2], 'k-')
axs[0, 0].set(xlabel='Palate rating', ylabel='Overall rating')
axs[0, 0].set_title(f'(R Value: {np.round(lr[2], 3)})')

lr, a1, a2, xmin, xmax = getLinearRegression('aroma')
axs[0, 1].scatter(df_ratings_BA['aroma'], df_ratings_BA['rating'], s=.5, alpha=.7, color='purple')
axs[0, 1].plot([xmin,xmax],[a1,a2], 'k-')
axs[0, 1].set(xlabel='Aroma rating', ylabel='Overall rating')
axs[0, 1].set_title(f'(R Value: {np.round(lr[2], 3)})')

lr, a1, a2, xmin, xmax = getLinearRegression('taste')
axs[1, 0].scatter(df_ratings_BA['taste'], df_ratings_BA['rating'], s=.5, alpha=.7, color='purple')
axs[1, 0].plot([xmin,xmax],[a1,a2], 'k-')
axs[1, 0].set(xlabel='Taste rating', ylabel='Overall rating')
axs[1, 0].set_title(f'(R Value: { np.round(lr[2], 3)})')

lr, a1, a2, xmin, xmax = getLinearRegression('appearance')
axs[1, 1].scatter(df_ratings_BA['appearance'], df_ratings_BA['rating'], s=.5, alpha=.7, color='purple')
lr, a1, a2, xmin, xmax = getLinearRegression('appearance')
axs[1, 1].plot([xmin,xmax],[a1,a2], 'k-')
axs[1, 1].set(xlabel='Appearance rating', ylabel='Overall rating')
axs[1, 1].set_title(f'(R Value: { np.round(lr[2], 3)})')

fig.tight_layout()
plt.show()
```
